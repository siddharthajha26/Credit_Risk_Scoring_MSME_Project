# We shall use here the Logistic Regression model and perform Model Validation and Performance Validation steps for clear understanding

**The ROC curve (Receiver Operating Characteristic curve) is a graph used to evaluate the performance of binary classification models across all possible decision thresholds**

The curve is created by plotting two key metrics derived from a model's confusion matrix: 

**True Positive Rate (TPR) / Sensitivity:** Plotted on the y-axis, this measures the proportion of actual positive cases correctly identified.(e.g defaults correctly classified as True Positives)

**False Positive Rate (FPR):** Plotted on the x-axis, this measures the proportion of actual negative cases incorrectly flagged as positive (e.g., non-default classified as default(FP)). It is calculated as 1 - Specificity

**Confusion matrix** Summarizes all correct and false predictions generated for a specific dataset
Key Interpretations: 

The Upper-Left Corner: The closer the curve is to the top-left coordinate (0,1), the more accurate the model is, as it represents high sensitivity with few false alarms.

The Diagonal Line: A 45-degree diagonal line from (0,0) to (1,1) represents a model with no predictive power, equivalent to random guessing.

Threshold Trade-offs: Every point on the curve represents a different classification threshold (e.g., 0.5 vs. 0.8). Moving along the curve illustrates how increasing sensitivity typically results in a higher false positive rate. 

**ROC AUC score** is the area under the ROC curve. It sums up how well a model can produce relative scores to discriminate between positive or negative instances across all classification thresholds. 
The ROC AUC score ranges from 0 to 1, where 0.5 indicates random guessing, and 1 indicates perfect performance.

https://www.evidentlyai.com/classification-metrics/explain-roc-curve

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
# We shall Create a UDF to plot the ROC curve,AUROC_score for the train,test and validation data.
def plot_auc_roc(data):
    y_true=data["default_flag"]
    y_hat=data["PD"]
    fpr, tpr, thresholds = roc_curve(y_true, y_hat)
    auc_value=roc_auc_score(y_true,y_hat)
    
    plt.figure(figsize=(15,8))
    plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {auc_value:.2f})')
    plt.plot(fpr, fpr, color='red', linestyle='--', label='Random Classifier')
    plt.xlabel("TPR")
    plt.ylabel("FPR")
    plt.title("ROC_Curve")
    plt.legend(loc='lower right')
    plt.show()
    return auc_value,print(f'(AUC :{auc_value: .2f})')
```

# Discriminatory Power
## GINI

GINI=2*(AUROC-1) measures how effectively a scoring model ranks borrowers by risk.The Gini score indicates the model's ability to concentrate "bad" loans (defaults) at the bottom of the predicted score range

0.0 (Random): The model is no better than a coin flip; defaults are spread randomly across all loan disbursement dates.

0.40.6 (Good/Standard): Typical for high-quality credit scorecards; the model effectively segments risk across different loan cohorts.

0.70.8+ (Excellent/Rare): High discriminatory power, often seen in behavior scorecards or models using extensive device data.
 


```python
def GINI(data):
    y_true=data["default_flag"]
    y_hat=data["PD"]
    AUC=roc_auc_score(y_true,y_hat)
    GINI=2*AUC-1
    return GINI
```

# CDF (Cummulative Distribution Function)
This function gives the probability a random variable X will be less than or equal to a specific value x, denoted as $(P(X\le x)$).
It works by accumulating probabilities, acting as a smooth or stair-step function from 0 to 1, telling you the cumulative likelihood of outcomes up to any given point, and is essential for understanding distributions, calculating probabilities over ranges (like $(P(a<X\le b)=F(b)-F(a)$)), and connecting to the Probability Density Function (PDF) for continuous variables.

**Continous Variables** $(F(x)$) is the integral (area under the curve) of the Probability Density Function (PDF) from negative infinity up to x.

**Discrete Variables**  $(F(x)$) is the sum (accumulation) of probabilities for all outcomes less than or equal to x.

**Properties:**

    -Always increases or stays flat (non-decreasing) as x increases.
    -Starts at 0 and ends at 1.
**How it's Used:**

    -Find probabilities in a range: $(P(a<X\le b)=F(b)-F(a)$).
    -Find probability of being above a value: $(P(X>x)=1-F(x)$).
    -Characterize distributions: Provides a complete picture of a random variable's behavior.

# Maximum Seperation 
## Kolmogorov-Smirnov (KS) statistic
The Kolmogorov-Smirnov (KS) statistic is a non-parametric metric that quantifies the maximum vertical distance between two cumulative distribution functions (CDFs). 
In binary classification, such as credit risk, it measures a model's ability to separate two classes (e.g., defaulters vs. non-defaulters)

The KS statistic ($(D$)) is defined as the supremum (maximum) absolute difference between two CDFs across all possible values of $(x$):$(D=\max |F_{1}(x)-F_{2}(x)|$)

**Empirical CDF (eCDF):** For real-world datasets, an empirical distribution is used. It is a step function that increases by $(1/n$) at each data point.
Concept: If two samples come from the same distribution, their eCDFs should "dance" close together. A large \(D\) indicates the samples likely originate from different distributions.

Scale: While ROC-AUC ranges from 0.5 to 1.0 for meaningful models, the KS statistic ranges from 0.0 to 1.0, making it more intuitive for non-technical stakeholders to interpret separation power

# Implementation Methodology (Binary Classification)
The standard implementation in machine learning involves comparing the distribution of predicted probabilities for positive and negative classes.

Observation-Level Method (Exact)
   
This method provides the most accurate unique value of KS. 

    -Step 1: Score your data using a predictive model to generate a probability of default (PD) for each observation.
    -Step 2: Sort all observations by their predicted scores in descending order.
    -Step 3: Calculate the Cumulative % of Positives (True Positive Rate/Recall) and Cumulative % of Negatives (False Positive Rate) at every observation point.
    -Step 4: Compute the difference at each point. The maximum difference found is the KS statistic.

```python
import numpy as np
from scipy.stats import ks_2samp

def calculate_ks_score(y_true, y_calibrated_probs):
    """
    Calculates the KS score for a set of calibrated probabilities.
    
    y_true: actual binary labels (0 or 1)
    y_calibrated_probs: calibrated probability scores for the positive class
    """
    # Separate scores by class
    scores_class_0 = y_calibrated_probs[y_true == 0]
    scores_class_1 = y_calibrated_probs[y_true == 1]
    
    # Calculate KS statistic (D) and p-value
    # D is the maximum difference between the two CDFs
    ks_statistic, p_value = ks_2samp(scores_class_0, scores_class_1)
    
    return ks_statistic


y_true = calibration_data["default_flag"]
y_probs = calibration_data["calibrated_pd"]
ks_score = calculate_ks_score(y_true, y_probs)
print(f"KS Score: {ks_score:.4f}")
```

```python
# Alternately, we can also use the eCDF code
import numpy as np
import matplotlib.pyplot as plt

def plot_ks_curve(y_true, y_probs):
    """
    Plots the KS curve for binary classification calibrated scores.
    """
    # Separate scores by class
    scores_0 = np.sort(y_probs[y_true == 0])
    scores_1 = np.sort(y_probs[y_true == 1])
    
    # Calculate empirical CDFs
    cdf_0 = np.arange(1, len(scores_0) + 1) / len(scores_0)
    cdf_1 = np.arange(1, len(scores_1) + 1) / len(scores_1)
    
    # Use a common set of points for interpolation to find the max difference
    all_points = np.sort(np.unique(np.concatenate([scores_0, scores_1])))
    cdf_0_interp = np.interp(all_points, scores_0, cdf_0, left=0)
    cdf_1_interp = np.interp(all_points, scores_1, cdf_1, left=0)
    
    # Calculate KS Statistic and its location
    diff = np.abs(cdf_0_interp - cdf_1_interp)
    ks_stat = np.max(diff)
    idx = np.argmax(diff)
    ks_x = all_points[idx]
    
    # Plotting
    plt.figure(figsize=(8, 6))
    plt.plot(all_points, cdf_0_interp, label='Class 0 (Negative)', color='blue', lw=2)
    plt.plot(all_points, cdf_1_interp, label='Class 1 (Positive)', color='green', lw=2)
    
    # Highlight the KS Statistic
    plt.vlines(ks_x, cdf_0_interp[idx], cdf_1_interp[idx], 
               colors='red', linestyles='--', label=f'KS Stat: {ks_stat:.3f}')
    
    plt.title('Kolmogorov-Smirnov (KS) Curve')
    plt.xlabel('Calibrated Score (Probability)')
    plt.ylabel('Cumulative Distribution')
    plt.legend(loc='lower right')
    plt.grid(alpha=0.3)
    plt.show()


y_true = calibration_data["default_flag"]     # Running KS statistics on the validation data with  calibrated_pd -----------------1
y_calibrated_probs = calibration_data["calibrated_pd"]
plot_ks_curve(y_true, y_calibrated_probs)
```

# Data-Drift
## Population Stability Index(PSI) and Stability Index(SSI)
It is a monitoring tool for ML models.PSI is measure of drift of model data(Features/Target/Predicted Variable) which occurs due to changing Production data that keeps evolving and growing with time.This index measures the drift in the model data over a training(Baseline) data and a new(Production) data.PSI compares the distribution of a key variable (like predicted scores or input features) in current data against the original training data.

**For Continuous Variables**

Binning: Data is typically divided into  bins (e.g., score ranges), and percentages of records in each bin are calculated for both the baseline and current data.

Formula: The formula sums up $($%$\text Actual - $%$ \text Expected) * ln($%$ \text Actual / $%$ \text Expected) $ for each bin, resulting in a single PSI value.
Interpretation:

    -PSI < 0.1: Stable, minimal change.
    -0.1 < PSI < 0.25: Moderate change, potentially needs monitoring.
    -PSI > 0.25: Significant change, model likely needs retraining. 
Why it's important? 

    -Model Monitoring: Detects "data drift" or "population shift" in real-world application.
    -Credit Risk: Ensures credit scoring models remain accurate as customer behavior changes.
    -Reliability: Helps maintain model performance and reliability over time, preventing inaccurate predictions due to changing input data.
**For Categorical Variables**
Unlike continuous variables, categorical variables do not require binning (discretizing) because the categories themselves act as natural "bins".

-Step 1: Identify the union of all categories ($(C)$) present in both the reference (Expected) and monitored (Actual) populations.
    
-Step 2: Calculate the proportion of records for each category C in both datasets:$(\text r_{c}$) (Expected %): Count of records in category C divided by the total reference population.
$(m_{c}$) (Actual %): Count of records in category C divided by the total monitored population.

-Step 3: Apply the formula for each category and sum the results:
$(\text PSI$=$ \sum _{c\in C}(\text r_{c}-m_{c})\times \ln \left(\frac{r_{c}}{m_{c}}\right)$)

**Limitations and Practical Implementation**

    -Cardinality: PSI is most effective for features with moderate cardinality (e.g., 50100 unique values). For variables with extremely high cardinality, the index may become less meaningful.
    -Grouping: If a variable has too many categories, you can group infrequent categories into an "Other" bucket to simplify monitoring.
    -Zero-Count Handling: If a category exists in one dataset but not the other, the natural log becomes undefined. To fix this, a small constant (e.g., 0.0001 or $(1e^{-6}$)) is typically added to the zero-count proportion.

```python
# The method of calculating PSI an SSI for all the features in the data would follow :

#- threshold: The "switch" that decides which of the two functions to use for a specific column.This will form the outerloop of the code where we shall 
# evaluate whether to apply PSI or SSI based on a threshold values of number of unique categories in a column.-----used in inner block

#- Calculate SSI(for categorical columns that have natural bins based on the number of unique values in the column)-----Inner block
#-calculate the PSI for numerical columns. First group numbers into ranges, then compare the percentages of those ranges.---inner block

# The code should also handle the cases where the new data has particular values which are absent in the training data and we shall formulate the code to
# handle these missing values.----first outer block
```

```python
import pandas as pd
import numpy as np
#---first outer block
def calculate_stability_score(expected_probs, actual_probs):
    """ The core PSI/SSI formula with zero-handling. """
    # Replace 0 with a tiny number to avoid math errors (division by zero / log(0))
    expected_probs = np.where(expected_probs == 0, 0.0001, expected_probs)
    actual_probs = np.where(actual_probs == 0, 0.0001, actual_probs)
    
    # Formula: (Actual % - Expected %) * ln(Actual % / Expected %)
    score = (actual_probs - expected_probs) * np.log(actual_probs / expected_probs)
    return np.sum(score)

def get_stability_report(train_df, test_df, threshold=10, bins=10):
    results = []
    
    for col in train_df.columns:
        # Step 1: Group the data (Binning for Numbers vs. Categories for Labels)
        if train_df[col].nunique() > threshold:  # theshold used as condition in the inner block
            # Numerical: Create 10 equal-width bins
            labels = pd.cut(train_df[col], bins=bins)
            expected = train_df[col].value_counts(bins=bins, normalize=True).sort_index()
            actual = pd.cut(test_df[col], bins=labels.cat.categories).value_counts(normalize=True).sort_index()
        else:
            # Categorical: Use the unique values as groups
            expected = train_df[col].value_counts(normalize=True)
            actual = test_df[col].value_counts(normalize=True)
            # Align categories so they match
            expected, actual = expected.align(actual, fill_value=0)

        # Step 2: Calculate the score
        score = calculate_stability_score(expected.values, actual.values)
        results.append({"Variable": col, "Stability_Score": score})

    return pd.DataFrame(results)
```

```python
train_df=train_data[['Net_Profit_Margin', 'ROA', 'sector_risk', 'ROE','saving_ratio']]
       
test_df=test_data[['Net_Profit_Margin', 'ROA', 'sector_risk', 'ROE','saving_ratio']]
       
expected_probs=test_data["PD"]
actual_probs=train_data["default_flag"]

print(get_stability_report(train_df, test_df, threshold=10, bins=10))
```

# Multi Collinearity
## Variance Inflation Factor (VIF) Calculation
**The Variance Inflation Factor (VIF)**
It is a statistical diagnostic used to detect multicollinearitya situation where independent variables in a regression model are highly correlated with one another.
1. Concept and ImportanceVIF quantifies how much the variance (uncertainty) of an estimated regression coefficient is "inflated" due to its relationship with other predictors.
  
**The Problem:** 
High multicollinearity makes it difficult to determine the individual impact of each variable on the target, leading to unstable coefficients and wider confidence intervals.
**The Formula:** 
For a given variable $(X_{i}$), $(VIF=\frac{1}{1-R_{i}^{2}}$), where $(R_{i}^{2}$) is the coefficient of determination obtained by regressing $(X_{i}$) against all other independent variables in the model.

Thresholds :

    #1: No correlation.
    
    #15: Moderate correlation generally acceptable.
    
    #> 5 or 10: High to serious multicollinearity; investigation or removal of variables is typically required.
**3. Remediation for High VIF**

If  variables with high VIF (e.g., > 5) are identified, consider these actions:

Variable Removal: Drop the feature with the highest VIF and recalculate the scores for the remaining features.

Feature Engineering: Combine highly correlated variables into a single representative feature.

Dimensionality Reduction: Use techniques like Principal Component Analysis (PCA) to create uncorrelated components.

Regularization: Switch to Ridge or Lasso regression, which penalize large coefficients caused by multicollinearity. 

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# 1. Load your independent variables (X)
X = train_data[['Net_Profit_Margin', 'ROA', 'sector_risk', 'ROE','saving_ratio']]


# 2. Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) 
                   for i in range(X.shape[1])]

# 3. View results (exclude the 'const' row for final interpretation)
print(vif_data)
```

# Concentration Risk
## Herfindahl-Hirschman Index (HHI) Calculation and Adjusted HHI



Herfindahl-Hirschman Index (HHI) is a vital tool used to measure Concentration Risk. Financial institutions use it to ensure they aren't "putting too many eggs in one basket"
Formula: $(HHI=\sum _{i=1}^{n}(w_{i})^{2}$), where $(w_{i}$) is the percentage of the total portfolio exposure belonging to borrower or sector $(i$)

**Ideal HHI Value Ranges for Loan Portfolios**

    #< 0.10 (or < 1,000 points): Highly Diversified. The portfolio is spread across many sectors or borrowers, indicating low concentration risk.
    #0.10  0.15 (1,000  1,500 points): Generally Diversified. Considered acceptable for most commercial banks.
    #0.15  0.25 (1,500  2,500 points): Moderately Concentrated. Risk levels are increasing, requiring monitoring.
    #> 0.25 (or > 2,500 points): Highly Concentrated. This indicates high risk, as a significant portion of the portfolio is in a few sectors/borrowers

The **Adjusted HHI** (or "Granularity Adjustment") is used to account for the fact that even if a portfolio is large, if the individual loan sizes are unequal, the risk is higher.Banks use this to adjust their Internal Ratings-Based (IRB) capital requirements. A high concentration (high HHI) suggests that the "Law of Large Numbers" isn't working effectively to smooth out defaults, so the bank must hold more capital.

Regulatory Capital: High HHI directly increases the Economic Capital a bank must hold under Pillar 2 of the Basel framework

Sector Limits: Banks use HHI to set "Hard Limits" for industries (e.g., limiting Real Estate exposure to ensure HHI stays below 1,500

Stress Testing: The Adjusted HHI is a standard input for stress test scenarios to determine how a portfolio would react if the largest "concentrated" borrowers defaulted simultaneously

**The Calculation Process:**

    -Calculate the total number of exposure for each borrower.
    -Portfolio Weight ($(w_{i}$)): Divide individual exposure by the total portfolio value.
    -HHI Score: Square and sum these weights.
    -Adjustment: If HHI is above the bank's internal threshold (often mapped to the 2,500 "High Concentration" mark), a capital surcharge is applied.

# Model Stability,Accuracy and Reliability
## Anchor Point Test
Anchor Point test typically refers to a calibration and benchmarking procedure used to ensure that a model's risk estimatessuch as Probability of Default (PD)are aligned with stable, long-run expectations or external industry benchmarks.

**Purpose** 

Stability: It anchors model outputs to a "Through-the-Cycle" (TTC) or long-run average to prevent volatile, procyclical shifts during short-term economic fluctuations.

Accuracy: It acts as a "reality check" by comparing internal model predictions against established external standards (e.g., S&P or Moodys ratings) to ensure risk is not underestimated.

Reliability: In low-data environments (like SME or corporate lending), it provides a "defendable" starting point for ratings by anchoring internal factors to broader market benchmarks. 

The test generally follows a three-step validation flow:

Selection of the Anchor: Validation teams identify a reliable reference point. 
**Common anchors include:**

Long-run Default Rates: An average default rate over at least one full economic cycle.

Central Tendency: The target average rating for a specific portfolio segment.

External Benchmarks: Globally accepted credit rating scales (e.g., AAA to D).

Calibration Comparison: The models predicted PD values are binned and compared against the observed default rates or the anchor's benchmark. If the model's "Point-in-Time" (PIT) estimates deviate significantly from the anchor without economic justification, it signals a need for recalibration.

Stability Analysis: Analysts monitor the Population Stability Index (PSI) and Weight of Evidence (WoE) trends relative to these anchor points to detect if the borrower composition is shifting away from the model's design.

    -Regulatory Compliance: Under Basel IRB frameworks, banks must prove their PDs are calibrated to a long-run anchor to ensure capital adequacy during downturns.
    -Accounting Standards: For IFRS 9/CECL, anchoring a "neutral long-run economic state" into forecasting filters helps reduce noise and produces smoother, more stable lifetime PD projections.

```python
import pandas as pd

def Anchor_point(data, PD, total_observations, central_tendency, upper_green=1.2, upper_red=1.3, lower_green=0.8, lower_red=0.7):
    """
    Perform Anchor Point test for a given master scale data.
    """
    # Calculate the weighted average PD across all rating grades
    # Weighted Average PD = Sum(PD_i * Count_i) / Sum(Count_i)
    total_obs_sum = data[total_observations].sum()
    weighted_pd_sum = (data[PD] * data[total_observations]).sum()
    avg_pd = weighted_pd_sum / total_obs_sum

    # Calculate thresholds based on the average PD
    upper_g = avg_pd * upper_green
    upper_r = avg_pd * upper_red
    lower_g = avg_pd * lower_green
    lower_r = avg_pd * lower_red

    # Create the results DataFrame
    results_df = pd.DataFrame({
        'central_tendency': [central_tendency],
        'avg_pd': [avg_pd],
        'lower_r': [lower_r],
        'lower_g': [lower_g],
        'upper_g': [upper_g],
        'upper_r': [upper_r]
    })

    # Apply traffic light logic
    def classify(row):
        ct = row['central_tendency']
        if row['lower_g'] <= ct <= row['upper_g']:
            return 'Green'
        elif ct < row['lower_r'] or ct > row['upper_r']:
            return 'Red'
        else:
            return 'Yellow'

    results_df['test_result'] = results_df.apply(classify, axis=1)

    return results_df

```

#  The Chi-Square  test 

It is a foundational statistical tool used primarily for variable selection, scorecard development, and model validation. It functions by comparing observed frequencies of defaults and non-defaults against expected frequencies to determine if a relationship is statistically significant rather than due to random chance.
## Core Applications 
**Feature Selection (Information Value):**
It is a standard "filter method" used to identify which categorical features (e.g., employment type or residence status) are most predictive of loan default. A higher chi-square score indicates a stronger association with the target variable.

**Optimal Binning (Chi-Merge):**

Used to discretize continuous variables (like age or income) by merging adjacent intervals that have the most similar default behavior, ensuring that each "bin" in a scorecard is statistically distinct.

**Goodness-of-Fit (Validation):** 

Analysts use the Hosmer-Lemeshow test (a specific Chi-Square test) to verify if the number of actual defaults in different risk grades matches the model's predicted Probability of Default (PD).

**Scorecard Stability:** 

It measures the difference between actual and predicted distributions across all borrower characteristics; lower chi-square values indicate a more stable scorecard.

**Calculation and Decision Framework**

The test uses the formula:$(\chi ^{2}=\sum \frac{(O-E)^{2}}{E}$)) 

(Observed): The actual number of "Bad" (defaulters) and "Good" borrowers in a specific category.
$(E$) (Expected): The number of "Bad" and "Good" borrowers expected if the category had no influence on credit risk.
Hypothesis Testing:Null Hypothesis ($(H_{0}$)): There is no association between the feature and credit risk (the variables are independent).
Alternative Hypothesis ($(H_{A}$)): The feature is significantly associated with credit risk.

Decision: If the calculated $(\chi ^{2}$) exceeds a critical value (or the p-value < 0.05), $(H_{0}$) is rejected, and the feature is kept for the model.

**Implementation ConstraintsCategorical Data Only:**
Continuous variables must be converted into bins/categories before the test can be applied.

Large Sample Requirement: For the test to be reliable in audits, each cell in the contingency table should have an expected frequency of at least 5.

Correlation vs. Causation: The test indicates an association exists (e.g., "self-employed" borrowers default more often) but does not prove that being self-employed causes the default. 

 

```python
# Perform a Chi Square test on master scale data to assess the goodness-of-fit between observed and expected bad rates.
# Calculates the expected bad rate (expected_bad) based on PD and total_observations.
# Computes the Chi Square statistic (chi_square) to measure the discrepancy between observed and expected bad rates.
# Calculates the p-value using the Chi Square statistic and degrees of freedom.
# Compares the p-value against the specified confidence_level to determine if the rating scale passes or fails the test.

import pandas as pd
import numpy as np
from scipy.stats import chi2


def chisquare_test(data, PD, observed_bad, total_observations, confidence_level=0.95):
    """
    Chi Square Test

    This function allows performing a chi-square test on master scale data.

    Parameters:
    data : pandas DataFrame
        A master scale data set.
    PD : str
        Name of the PD variable.
    observed_bad : str
        Name of the observed defaults variable.
    total_observations : str
        Name of the total observations variable.
    confidence_level : float, optional
        Confidence level for the test, default is 0.95.

    Returns:
    dict
        A dictionary containing the following keys:
        - 'data': DataFrame with calculated fields
        - 'p_value': The p-value of the chi-square test
        - 'result': The result of the test

    Examples:
    chisquare_test(master_scale_data, "PD", "Bad_obs", "Tot_obs", 0.90)
    """
    data['expected_bad'] = data["Total.Observations"] * data[PD]
    data['chi_square'] = (((data['expected_bad'] - data[observed_bad]) ** 2) / data['expected_bad'])
    chi2_statistic = np.sum(((data['expected_bad'] - data[observed_bad]) ** 2) / data['expected_bad'])
    p_value = 1 - chi2.cdf(chi2_statistic, df=1)  # Calculating p-value
    cl = 1 - confidence_level
    if p_value > cl:
        result = f"The rating scale did not pass the test {round(p_value, 3)} > {cl}"
    else:
        result = f"The rating scale passed the test. {round(p_value, 3)} < {cl}"

    return {
        'data': data,
        'p_value': p_value,
        'result': result
    }
```

```python
# Chi-square test
chisquare_result = chisquare_test(master_scale_data, "PD", "Bad.Count", "Total.Observations", 0.90)
print("Chi-square Test Result:", chisquare_result)
```

# Back Testing
## Binomial Test

The Binomial Test is a standard backtesting tool used to validate the calibration of Probability of Default (PD) models. It determines whether the observed number of defaults in a specific rating grade is statistically consistent with the model's predicted PD.
**Core Concept**

In Credit RiskThe test treats each loan in a portfolio as a Bernoulli trial (it either defaults or it doesn't). 
In a rating grade with (N) borrowers and a predicted probability of default (P), the number of observed defaults (K) should theoretically follow a binomial distribution $(K\sim \text{Binomial}(N,P)$).

**One-tail Test (Basel IRB):**  Typically used for conservatism. It tests if the model underestimates risk (i.e., actual defaults are significantly higher than expected).

**Two-tail Test (IFRS 9):** Used for accuracy to ensure the PD estimate is neither too high nor too low.

**Implementation Steps**
To perform a binomial test on a credit portfolio, follow these logical steps:
1-Define Hypotheses:Null Hypothesis ($(H_{0}$)): The predicted PD is correct.
Alternative Hypothesis ($(H_{A}$)): The predicted PD underestimates the actual risk.Calculate the p-value:Calculate the probability of observing at least $(k$) defaults if $(H_{0}$) were true.
$(P(K\ge k)=\sum _{i=k}^{N}{N \choose i}P^{i}(1-P)^{N-i}$)

Evaluate Significance:Compare the p-value against a confidence level (usually 0.05 or 0.01). If the p-value is lower than the threshold, reject $(H_{0}$), indicating the model needs recalibration.

```python
import pandas as pd
from scipy.stats import norm

def Binomial_test(rating_scale_data, total_observations, PD, DR, confidence_level=0.90, tail="one"):
    """
    Perform binomial test on master scale data.

    Parameters:
    - rating_scale_data (DataFrame): Master scale data set.
    - total_observations (str): Total observations variable.
    - PD (str): PD variable.
    - DR (str): Default rate variable.
    - confidence_level (float): Confidence level, default is 0.90.
    - tail (str): Tail preference of the test, 'one' or 'two', default is 'one'.

    Returns:
    DataFrame: DataFrame containing test results.
    """
    def one_tail_test(data, total_col, pd_col, dr_col, conf_level):
        data['BadObs'] = data[total_col] * data[dr_col]
        data['BadEstimations'] = data[total_col] * data[pd_col]
        data['TestEstimation'] = data[pd_col] * data[total_col] + norm.ppf(conf_level) * \
                                 (data[pd_col] * data[total_col] * (1 - data[pd_col])) ** 0.5

        dif_TestEstimation = data['TestEstimation'] - data['BadObs']
        data['Test_Result'] = pd.cut(dif_TestEstimation, bins=[-float('inf'), 0, float('inf')],
                                      labels=["Target Value Underestimated", "Target Value Correct"],
                                      include_lowest=True, right=False)
        return data

    def two_tail_test(data, total_col, pd_col, dr_col, conf_level):
        data['BadObs'] = data[total_col] * data[dr_col]
        data['BadEstimations'] = data[total_col] * data[pd_col]
        data['TestEstimationUpper'] = data[pd_col] * data[total_col] + norm.ppf(conf_level) * \
                                      (data[pd_col] * data[total_col] * (1 - data[pd_col])) ** 0.5
        data['TestEstimationLower'] = data[pd_col] * data[total_col] - norm.ppf(conf_level) * \
                                      (data[pd_col] * data[total_col] * (1 - data[pd_col])) ** 0.5
        data['Test_Result'] = pd.cut(data['BadObs'], bins=[-float('inf'), data['TestEstimationLower'],
                                                           data['TestEstimationUpper'], float('inf')],
                                      labels=["Target Value Overestimated", "Target Value Correct",
                                              "Target Value Underestimated"], include_lowest=True, right=False)
        return data

    if tail == "one":
        return one_tail_test(rating_scale_data, total_observations, PD, DR, confidence_level)
    elif tail == "two":
        return two_tail_test(rating_scale_data, total_observations, PD, DR, confidence_level)
    else:
        print("Tail can only be 'one' or 'two'.")
```

```python
binomial_result = Binomial_test(master_scale_data, "Total.Observations", "PD", "Bad.Rate", 0.90, "one")
print("Binomial Test Result:", binomial_result)
```

## Adjusted Binomial Test

**Calculation Details**:
   - **One-Tail Test**:
     - Calculates the adjusted result for underestimation of bad observations.
     - Uses PD, default rate, confidence level, and default correlation (`r`) to estimate the lower bound of bad observations.
     - Categorizes the test result based on whether the estimated bad observations are correct or underestimated.

   - **Two-Tail Test**:
     - Calculates the adjusted results for overestimation and underestimation of bad observations.
     - Uses PD, default rate, confidence level, and default correlation (`r`) to estimate the upper and lower bounds of bad observations.
     - Categorizes the test result based on whether the observed bad observations fall within the estimated bounds.

```python
import pandas as pd
from scipy.stats import norm
from math import sqrt

def Adjusted_Binomial_test(rating_scale_data, total_observations, PD, DR, confidence_level=0.90, tail="one", r=0.40):
    """
    Perform adjusted binomial test on master scale data.

    Parameters:
    - rating_scale_data (DataFrame): Master scale data.
    - total_observations (str): Column name for total observations.
    - PD (str): Column name for PD variable.
    - DR (str): Column name for Default Rate variable.
    - confidence_level (float): Confidence level, default is 0.90.
    - tail (str): Tail preference of the test, 'one' or 'two', default is 'one'.
    - r (float): Default correlation between rating grades, default is 0.40.

    Returns:
    DataFrame: DataFrame with adjusted test results.
    """

    def one_tail_test(data, total_col, pd_col, dr_col, conf_level, r):
        data['BadObs'] = data[total_col] * data[dr_col]
        data['BadEstimations'] = data[total_col] * data[pd_col]
        data['Default_Correlation'] = r

        t = norm.ppf(data[pd_col])
        Q_param = norm.cdf((sqrt(r) * norm.ppf(conf_level) + t) / sqrt(1 - r))
        Adjusted_result = Q_param + 1 / (2 * data[total_col]) * (
                2 * Q_param - 1 + (Q_param * (1 - Q_param)) /
                norm.pdf((sqrt(r) * norm.ppf(1 - conf_level) - t) / sqrt(1 - r)) *
                ((2 * r - 1) * norm.ppf(1 - conf_level) - t * sqrt(r)) / sqrt(r * (1 - r)))

        data['TestEstimation'] = round(Adjusted_result * data[total_col], 2)

        dif_TestEstimation = data['TestEstimation'] - data['BadObs']
        data['Test_Result'] = pd.cut(dif_TestEstimation,
                                     bins=[-float('inf'), 0, float('inf')],
                                     labels=["Target Value Underestimated", "Target Value Correct"],
                                     include_lowest=True, right=False)
        return data

    def two_tail_test(data, total_col, pd_col, dr_col, conf_level, r):
        data['BadObs'] = data[total_col] * data[dr_col]
        data['BadEstimations'] = data[total_col] * data[pd_col]
        data['Default_Correlation'] = r

        t = norm.ppf(data[pd_col])
        Q_param = norm.cdf((sqrt(r) * norm.ppf(conf_level) + t) / sqrt(1 - r))
        Adjusted_resultUpper = Q_param + 1 / (2 * data[total_col]) * (
                2 * Q_param - 1 + (Q_param * (1 - Q_param)) /
                norm.pdf((sqrt(r) * norm.ppf(1 - conf_level) - t) / sqrt(1 - r)) *
                ((2 * r - 1) * norm.ppf(1 - conf_level) - t * sqrt(r)) / sqrt(r * (1 - r)))
        Adjusted_resultLower = Q_param + 1 / (2 * data[total_col]) * (
                2 * Q_param - 1 - (Q_param * (1 - Q_param)) /
                norm.pdf((sqrt(r) * norm.ppf(1 - conf_level) - t) / sqrt(1 - r)) *
                ((2 * r - 1) * norm.ppf(1 - conf_level) - t * sqrt(r)) / sqrt(r * (1 - r)))

        data['TestEstimationUpper'] = round(Adjusted_resultUpper * data[total_col], 2)
        data['TestEstimationLower'] = round(Adjusted_resultLower * data[total_col], 2)

        data['Test_Result'] = pd.cut(data['BadObs'],
                                     bins=[-float('inf'), data['TestEstimationLower'],
                                           data['TestEstimationUpper'], float('inf')],
                                     labels=["Target Value Overestimated", "Target Value Correct",
                                             "Target Value Underestimated"],
                                     include_lowest=True, right=False)
        return data

    if tail == "one":
        return one_tail_test(rating_scale_data, total_observations, PD, DR, confidence_level, r)
    elif tail == "two":
        return two_tail_test(rating_scale_data, total_observations, PD, DR, confidence_level, r)
    else:
        print("Tail can only be 'one' or 'two'.")
```

```python
adj_binomial_test_result = Adjusted_Binomial_test(master_scale_data,  "Total.Observations", "PD", "Bad.Rate", 0.90, "one", r=0.40)
print("Adjusted Binomial Test Result:", adj_binomial_test_result)
```
